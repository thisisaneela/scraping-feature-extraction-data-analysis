```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
import spacy
from spacy import displacy
nlp= spacy.load("en_core_web_sm")
```

```python
df=pd.read_csv("Feature Extraction.csv")
```

```python
df.head()
```

```python

# Group by 'Negative Sentiments' and 'Positive Sentiments' columns and calculate the mean for each group
grouped_data = df.groupby(['Negative Sentiments', 'Postive Sentiments']).mean().reset_index()

plt.figure(figsize=(7, 7))
sns.barplot(data=grouped_data, x='Negative Sentiments', y='Postive Sentiments', palette='rocket')
plt.show()
```

```python
import pygal
from IPython.display import SVG

# Replace 'Negative Sentiments', 'Positive Sentiments', and 'Neutral Sentiments' with the actual column names from your DataFrame

chart = pygal.HorizontalBar()
chart.title = 'Long, medium, and short routes'
chart.add('Long', df['Negative Sentiments'])
chart.add('Medium', df['Postive Sentiments'])
chart.add('Short', df['Neutral Sentiments'])

# # Render the chart as SVG and display it
chart.render()
SVG(chart.render())


# Save the chart as an image file (e.g., PNG)
chart.render_to_file('chart.png')




```
